```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sg
```

```python
turndata = np.genfromtxt("Data/test4_right_turns.csv", delimiter=",", skip_header=1)
turbdata = np.genfromtxt("Data/test2_bumpy_road.csv", delimiter=",", skip_header=1)
```

```python
tdata = np.transpose(tdata[[0,1,2,4],:])
```

```python
accfig = plt.figure(figsize=(9,5))
plt.plot(turbdata[:,0], turbdata[:,4]/9.8)
plt.plot(turndata[:,0], turndata[:,4]/9.8)
plt.legend(["Turbulence (Bumpy Road)", "Right Turns"])
plt.title("Total Acceleartion from Turbulence and Turns")
plt.xlabel("Seconds")
plt.ylabel("Acceleration (g)")
plt.show()
```

## Filtering

```python
fs = 1/(turbdata[1,0]-turbdata[0,0])
sos = sg.butter(10, 3, fs=fs, output='sos')
```

```python
filtered_turns = sg.sosfilt(sos, turndata[:,4]/9.8)
filtered_turb = sg.sosfilt(sos, turbdata[:,4]/9.8)
ffig = plt.figure(figsize=(9,5))
plt.plot(turbdata[:,0], filtered_turb)
plt.plot(turndata[:,0], filtered_turns)
plt.show()
```
